# Step 0. xlsx to csv

```python
import pandas as pd
import os
from pathlib import Path
```

```python
# ./xlsx에 위치한 폴더를 순회하며 dir 구조 파악

xlsx_files = []
input_folder = './xlsx'

for dirpath, dirnames, filenames in os.walk(input_folder):
    for filename in filenames:
        if filename.endswith(".xlsx"):
            xlsx_files.append(Path(dirpath) / filename)

xlsx_files
```

```python
# xlsx_files에 있는 모든 xlsx 파일을 읽어서 각각 csv로 저장
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")

for xlsx_file in xlsx_files:
    # xlsx 파일 읽기
    df = pd.read_excel(xlsx_file, engine='openpyxl')
    
    # 출력 폴더 경로 설정 (입력 폴더와 동일한 위치에 저장)
    output_folder = Path('./xlsx_to_csv') / xlsx_file.parent.relative_to(input_folder)
    output_folder.mkdir(parents=True, exist_ok=True)
    output_file = output_folder / (xlsx_file.stem + '.csv')
    
    # CSV로 저장
    df.to_csv(output_file, index=False, encoding='utf-8-sig')
    
    print(f"Converted {xlsx_file} to {output_file}")
```
